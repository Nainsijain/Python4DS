## Creating new feature

```python
df['title'][:10]
```

```python
def make_reason(x):
    x=x.split(':')[0]
    return x
df['Reason']= df['title'].apply(make_reason)
df.head()
```

```python
# most common reason for 911 calls based on that new column
df['Reason'].value_counts().index[0]
```

```python
# display unique titles for Reason='EMS'
df[df['Reason']=='EMS']['title'].unique()
```

```python
# plotting the reason column
sns.countplot(df['Reason'])
```

## DateTime column
timeStamp

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'][0])
```

```python
df['Hour']= df['timeStamp'].apply(lambda t: t.hour)
df['Month']= df['timeStamp'].apply(lambda t: t.month)
df['Day of Week']= df['timeStamp'].apply(lambda t: t.dayofweek)
```

```python
df.head()
```

```python
d= {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week']= df['Day of Week'].map(d)
```

```python
df.head(1)
```

```python
# number of calls based on day of week
sns.countplot(df['Day of Week'], hue=df['Reason'])
```

```python
sns.countplot(df['Month'], hue=df['Reason'])
```

```python
byMonth=df.groupby('Month').agg('count')
byMonth
```

```python
byMonth['lat'].plot()
```

## lmplot()

```python
byMonth
```

```python
byMonth.reset_index(inplace=True)
byMonth
```

```python
sns.lmplot(data=byMonth, x='Month', y='lat')
```

```python
df['Date']= df['timeStamp'].apply(lambda x: x.date())
```

```python
df['Date'].head(2)
```

```python
df[df['Reason']=='Fire'].groupby('Date').agg('count')['lat'].plot(label='Fire')
df[df['Reason']=='EMS'].groupby('Date').agg('count')['lat'].plot(label='EMS')
df[df['Reason']=='Traffic'].groupby('Date').agg('count')['lat'].plot(label='Traffic')
plt.legend()
```
